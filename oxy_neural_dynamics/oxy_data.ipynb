```python
from scipy.io import loadmat
oxy = loadmat('../data/O2_internal_20220806.mat')['O2_internal'].T
low_, high_ = np.percentile(oxy, [5, 95], axis=1)
oxy_ = (oxy - low_[:, None])/(high_ - low_)[:, None]
```

```python
np.savez('../data/O2_internal.npz', oxy_ = oxy_, oxy_mean = oxy_.mean(axis=0))
```

```python
plt.plot(oxy_.T, lw=0.5)
plt.xlim([0, 3600])
plt.vlines([1200, 2400], ymin=0, ymax=1.5, colors='k', linestyles='--')
plt.xlabel('Time (s)')
plt.ylabel('Normalized oxygen level')
sns.despine()
plt.show()
```

```python
plt.plot(oxy_.mean(axis=0), '-k', lw=0.5)
plt.vlines([1200, 2400], ymin=0, ymax=1.0, colors='k', linestyles='--')
plt.xlim([0, 3600])
plt.xlabel('Time (s)')
plt.ylabel('Normalized oxygen level')
sns.despine()
plt.show()
```
